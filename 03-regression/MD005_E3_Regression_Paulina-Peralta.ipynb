# **Bienvenid@ al PokéLab - Regresión**

En este cuaderno entrenaremos modelos de regresión, para descubrir si con los datos de una "Pokédex" podemos predecir estadísticas de combate de un Pokémon.

## Dataset: Pokédex

El dataset utilizado es "Data of 1015 Pokémons" de Kaggle, que funciona como una Pokédex estructurada, el cual incluye información de 1025 Pokémons con variables de identidad, tipos, estadísticas de combate, rasgos y extras.

#### Identidad
- `id` — número de Pokédex  
- `name` — nombre del Pokémon  
- `rank` — rareza/categoría (ordinary, legendary, mythical, sub-legendary, etc.)  
- `generation` — región/generación (generation-i, generation-ii, …)  
- `evolves_from` — evolución previa  

#### Tipos
- `type1` — tipo principal  
- `type2` — tipo secundario (puede ser `None`)  

#### Stats de combate
- `hp`, `atk`, `def`, `spatk`, `spdef`, `speed` — estadísticas base  
- `total` — suma total de stats  

#### Rasgos y extras
- `height`, `weight` — medidas físicas  
- `abilities` — habilidades  
- `desc` — descripción Pokédex

### Variables Numéricas

En primer lugar separamos y analizamos las variables numéricas del dataset, de forma a comprender su distribución y entender cuales features nos ayudarán a diseñar el modelo de regresión.

```python
numeric_cols = ['hp', 'atk', 'def', 'spatk', 'spdef', 'speed', 'total', 'height', 'weight']
print(f"Variables numéricas: {numeric_cols}")
dataset[numeric_cols].describe()
```

#### **Histogramas**

A continuación gráficamos las distribuciones de las variables numéticas mediante histogramas, para entender el comportamiento de cada una de las variables, visualizando la media, los picos, la variabilidad, outliers, etc.

```python
fig, axes = plt.subplots(3, 3, figsize=(14, 10))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    axes[i].hist(dataset[col], bins=30, color='#5060e1', edgecolor='white', alpha=0.7)
    axes[i].set_title(f'Distribución de {col}', fontweight='bold')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frecuencia')
    axes[i].axvline(dataset[col].mean(), color='red', linestyle='--', label=f'Media: {dataset[col].mean():.1f}')
    axes[i].legend(fontsize=8)

plt.tight_layout()
plt.suptitle('Distribución de Variables Numéricas', fontsize=14, fontweight='bold', y=1.02)
plt.show()
```

#### Test de Normalidad Shapiro-Wilk

Realizamos el test de Shapiro-Wilk para verificar si las variables siguen una distribución normal, lo cual es importante para algunos modelos de regresión. Para ello hacemos dos hipótesis:

- **H₀ (Hipótesis nula):** Los datos siguen una distribución normal
- **H₁ (Hipótesis alternativa):** Los datos NO siguen una distribución normal

Si el test de Shapiro-Wilk es cercano a 1, la distribución es más compatible con la normalidad y si es cercano a cero es menos compativle con la normalidad.

Por lo cual, si:
- p > 0.05: No se rechaza la hipótesis nula, ya que no podemos asumir normalidad.
- p ≤ 0.05: Se rechaza la hipótesis nula.

```python
for col in numeric_cols:
    data = dataset[col].dropna()
    stat, p_value = stats.shapiro(data)
    print(f"{col}: estadístico={stat:.4f}, p-valor={p_value:.4f}")
```

Para todas las variables el p-valor del test de Shapiro-Wilk es 0.05<, por lo tanto no se cumple el supuesto de normalidad, entonces rechazamos la hipótesis nula. Lo cual es esperable ya que las estadísticas de Pokémon siguen distribuciones específicas del dominio. Lo cual significa que para el diseño del modelo de regresión, modelos como Random Forest que no asumen normalidad podrían ser más adecuados.

Generalmente los Pokémones evolucionados mejoran sus estadísticas a excepcion de algunos que no evolucionan como por ejemplo Mewtwo que es un Pokémon legendario. Para analizarlos mejor cree una variable binaria de evolución.

```python
dataset['evolves'] = (dataset['evolves_from'] != 'nothing').astype(int)
```

Y cree una funcion que determina la etaá evolucitva del Pokémon, para verificar si el Pokémon del que evoluciona, tiene una evolución también.

```python
def count_evolution_stage(row):
    """Determina la etapa evolutiva del Pokémon"""
    if row['evolves_from'] == 'nothing':
        return 'Base'
    else:
        parent = row['evolves_from']
        parent_data = dataset[dataset['name'] == parent]
        if len(parent_data) > 0 and parent_data.iloc[0]['evolves_from'] != 'nothing':
            return 'Stage 3'
        else:
            return 'Stage 2'
dataset['evolution_stage'] = dataset.apply(count_evolution_stage, axis=1)
```

### Análisis de multicolelinealidad

Para una correcta clasificación debemos analizar antes si hay variables con alta correlación, ya que si las hay podría haber multicolinealidad, que significa que las variables estan muy relacionadas entre sí, y hace que el modelo no pueda separar el efecto único de cada variable, generando coeficientes intestables y difíciles de interpretar. Para ello correlacionamos todas las features incluyendo a la variable objetivo y la graficamos en la matriz de correlación.

```python
corr_cols = pd.concat([X_scaled, dataset[['atk']]], axis=1)
correlation_matrix = corr_cols.corr()
plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='RdBu_r', center=0,fmt='.2f', square=True, linewidths=0.5)
plt.title('Matriz de Correlación con la variable objetivo', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

En la matriz de correlación se visualiza que la variable total es la más relacionada con atk la variable objetivo con 0.71 lo cual tiene sentido por como mencionamos antes la variable total es una variable compuesta de las estadísticas de combate de los Pokémones por lo tanto no podremos utilizarla para el análisis.

Las variables codificadas type1_encoded, generation_encoded y has_type2 no tienen mucha correlación con las demás variables lo que podría indicar que estas aportan otro tipo de información.

Luego analizando las features agregadas max_stat tiene una alta correlación con stat_range, stat_std, stats_above_100, y total_percentile por lo tanto podría ser redundante para el modelo. min_stat tiene una alta correlación con total_percentile y stats_above_100. stat_range tiene una alta correlación con stat_std. De estas variables decido dejar max_stat y min_stat porque tienen mayor correlación con la variable objetivo y no presentan una correlación tan alta entre sí, lo que indica que estas agregan una información extra al modelo. Y decidimos quitar: stat_std, stats_above_100, stat_range. La variable creada hp_x_def es la que muestra mayor correlación con la cariable objetivo atk, pero muestra también alta correlación con hp, y def,por lo tanto para que no haya multicolinealidad tendremos en cuenta solo hp_x_def.

Luego height y weight tienen una correlacion super alta de 0.63, por lo cual estas variables podrían tener la misma información. Donde la variable weight es la que mas se correlaciona con atk por lo tanto decido dejar está y quitar del análisis a height.

Si analizamos las correlaciones con la variable objetivo, se observa que las variables con mayor correlación positiva son: hp, min_stat, max_stat, def y weight. Lo cual es interesante porque significa que hay cierta relación entre los puntos de salud y las demás estadísticas de combate en el la variable de ataque.

```python
correlations_significant = []
for col in X_scaled.columns:
    corr, p_value = pearsonr(corr_cols[col], corr_cols['atk'])
    correlations_significant.append({
        'variable': col,
        'correlacion': corr,
        'p_value': p_value,
        'significativa': p_value < 0.05
    })
df_corr_sig = pd.DataFrame(correlations_significant).sort_values('correlacion', key=abs, ascending=False)
print(df_corr_sig[df_corr_sig['significativa']])
```

La variable que figura como más significativa para el análisis es la variable total, la cual no la podremos considerar debido a que es una variable compuesta que tiene información de ataque, ya que representa a la suma de todas las variables de combate. **height y weight** tiene una alta correlación, la cual decidimos dejar weight porque presenta una correlación positiva más alta con la variable objetivo. **stat_range, stat_std y stats_above_100** tampoco las seguiremos considerando ya que tienen una alta correlación entre sí, y consideraremos min_stat y max_stat envez de estas.

### Función de evaluación

Creamos una funcion llamada evaluate_model la cual utilizaremos para hacer las predicciones, calcular las métricas, imprimir los resultados y la matriz de confusión para cada uno de los modelos que desarrollaremos en esta sección.

```python
def evaluate_regression_model(model, X_train, X_test, y_train, y_test, model_name):

    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)
    
    metrics = {
        'MAE (Train)': mean_absolute_error(y_train, y_pred_train),
        'MAE (Test)': mean_absolute_error(y_test, y_pred_test),
        'MSE (Test)': mean_squared_error(y_test, y_pred_test),
        'RMSE (Test)': np.sqrt(mean_squared_error(y_test, y_pred_test)),
        'R² (Train)': r2_score(y_train, y_pred_train),
        'R² (Test)': r2_score(y_test, y_pred_test)
    }
    
    print(f"{'='*50}")
    print(f"   {model_name}")
    print(f"{'='*50}")
    for metric, value in metrics.items():
        print(f"   {metric}: {value:.4f}")
    
    return metrics, y_pred_test
```

### Introducción a la búsqueda de hiperparámetros

<div align="center">

</div>

Optuna es un framework que utiliza algoritmos de optimización bayesiana, específicamente TPE (Tree-structured Parzen Estimator), que lo que hace es aprender de los resultados de las evaluaciones anteriores para sugerir mejores combinaciones de hiperparámetros en cada iteración. Para esto construye modelos probabilísticos que estiman qué valores de hiperparámetros tienen mayor probabilidad de producir buenos resultados basándose en los experimentos ya realizados.

Durante la optimización, Optuna crea un estudio (study) que va probando diferentes combinaciones de hiperparámetros y guarda los resultados de cada evaluación. El algoritmo TPE lo que hace es analizar los resultados anteriores y los divide en dos grupos: los que dieron buenos resultados y los que dieron malos resultados. Luego construye modelos probabilísticos para cada grupo y sugiere nuevos valores de hiperparámetros que tienen mayor probabilidad de pertenecer al grupo de buenos resultados.

Para utilizar Optuna necesitamos entender sus componentes principales:

- **Trial:** representa una evaluación individual del modelo con una combinación específica de hiperparámetros.

- **Study:** es el objeto principal que organiza y coordina los múltiples trials. 

- **Sampler:** es el algoritmo que decide qué hiperparámetros probar en cada trial. 

- **Objective function:** es la función que recibe un trial, construye y evalúa el modelo con los hiperparámetros sugeridos, y retorna la métrica a optimizar.

| Ventaja | Descripción |
|---------|-------------|
| **Eficiencia** | Encuentra buenos hiperparámetros con menos evaluaciones que GridSearchCV |
| **Flexibilidad** | Permite definir espacios de búsqueda complejos con diferentes tipos de distribuciones |
| **Escalabilidad** | Puede paralelizar las evaluaciones y utilizar técnicas de pruning |

Para optimizar los hiperparámetros de cada modelo de regresión elegido utilizaremos Optuna.

### Ridge Regression Polinómica

La regresión Ridge polinómica es un algoritmo de regresión que lo que hace es combinar transformaciones polinómicas de las features con regularización L2 para predecir valores continuos. Para esto utiliza PolynomialFeatures que transforma las features originales en combinaciones polinómicas, y luego aplica Ridge Regression que es un modelo lineal con regularización.

Durante el entrenamiento, primero transforma las features creando términos polinómicos que capturan relaciones no lineales entre las variables. Luego ajusta los coeficientes de estos términos polinómicos utilizando una función de pérdida que incluye la suma de los cuadrados de los coeficientes multiplicada por un parámetro alpha, lo cual previene el sobreajuste penalizando coeficientes muy grandes.

La regularización L2 lo que hace es penalizar la suma de los cuadrados de los coeficientes. Básicamente calcula la función de pérdida haciendo la sumatoria del cuadrado de los coeficientes, lo cual ayuda a mantener los coeficientes pequeños y estables, especialmente cuando hay multicolinealidad entre las features.

**Parámetros principales:**
- `poly_degree`: Grado del polinomio que determina hasta qué nivel se crean combinaciones polinómicas.
- `interaction_only`: Si es True, solo crea términos de interacción entre features, sin potencias individuales.
- `alpha`: Controla la fuerza de la regularización. Valores más altos penalizan más los coeficientes grandes, reduciendo el sobreajuste pero potencialmente aumentando el sesgo.
- `solver`: Algoritmo de optimización que resuelve el problema de minimización. Opciones como 'cholesky', 'svd', 'lsqr' que buscan los valores de los coeficientes que minimizan la función de pérdida.

#### Búsqueda de hiperparámetros

Para hacer una búsqueda eficiente con Optuna creamos una función llamada objective_ridge_poly que recibe un objeto trial de Optuna que contiene sugerencias de hiperparámetros que introducimos, de forma a que construya un pipeline con PolynomialFeatures y Ridge con esos hiperparámetros sugeridos, y lo evalue mediante validación cruzada con KFold configurado con 5 folds. Considerando que Optuna minimiza por defecto el cross_val_score con scoring='neg_mean_absolute_error' que retorna valores negativos, retornamos el negativo del promedio para que Optuna minimice correctamente el MAE.

Para la optimización configuramos optuna.create_study con direction='minimize' para minimizar el MAE, y sampler=TPESampler(seed=RANDOM_STATE) que es el algoritmo de muestreo que utiliza modelos probabilísticos para sugerir mejores hiperparámetros basándose en los resultados anteriores. Luego ejecutamos study_ridge_poly.optimize con n_trials=300 que indica que probará 300 combinaciones diferentes de hiperparámetros, y show_progress_bar=True para visualizar el progreso de la búsqueda en tiempo real.

```python
def objective_ridge_poly(trial):
    degree = trial.suggest_int('poly_degree', 1, 4)                                    #rango de valores polinomicos sugeridos
    interaction_only = trial.suggest_categorical('interaction_only', [True, False])    #para probar opciones de intereacciones entre features
    alpha = trial.suggest_float('alpha', 1e-4, 1e4, log=True)                          #para controlar la regularizacion L2 con log para buscar valores en escala log
    solver = trial.suggest_categorical('solver', ['auto', 'svd', 'cholesky', 'lsqr'])  #sugerencia de algoritmos de minimizacion
    
    model = Pipeline([
        ('poly', PolynomialFeatures(degree=degree, interaction_only=interaction_only, include_bias=False)),
        ('ridge', Ridge(alpha=alpha, solver=solver, random_state=RANDOM_STATE))
    ])
    
    cv = KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
    cv_scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='neg_mean_absolute_error', n_jobs=-1)
    
    return -cv_scores.mean()

study_ridge_poly = optuna.create_study(direction='minimize', sampler=TPESampler(seed=RANDOM_STATE))
study_ridge_poly.optimize(objective_ridge_poly, n_trials=1000, show_progress_bar=True)
```

Luego de la búsqueda iterativa de parametros con Optuna que probó 1000 hiperparámetros, el trial 274 fue el mejor consiguiendo un MAE medio de 17.1398 en la validación cruzada con datos de train.

#### Entrenar el modelo con los mejores parámetros encontrados

Luego entrenamos el modelo con los mejores hiperparámetros encontrados por Optuna. Para ello contruimos un pipeline final con los hiperparámetros óptimos encontrados.

```python
best_params_ridge = study_ridge_poly.best_params
best_ridge_poly = Pipeline([
    ('poly', PolynomialFeatures(degree=best_params_ridge['poly_degree'], interaction_only=best_params_ridge['interaction_only'], include_bias=False)),
    ('ridge', Ridge(alpha=best_params_ridge['alpha'], solver=best_params_ridge['solver'], random_state=RANDOM_STATE))
])
best_ridge_poly.fit(X_train, y_train)
print(f"\n Mejor MAE (Ridge Polinómica): {study_ridge_poly.best_value:.4f}")
print(f"Mejores parámetros: {study_ridge_poly.best_params}")
```

Utilizando como modelo un polinomio de segundo orden con una regularización de 13.76 aprox. y utilizando como solver svd, es un algoritmo que resuelve el sistema lineal de la regresión ridge descomponiendo la matriz de features en tres matrices mediante descomposición en valores singulares.

### Random Forest Regressor

Random Forest es un algoritmo de regresión de ensemble que lo que hace es construir múltiples árboles de decisión y promediar sus predicciones para obtener una estimación más robusta y precisa. Para esto utiliza el principio de "wisdom of the crowd", donde cada árbol se entrena con una muestra aleatoria diferente del dataset (bootstrap) y considera solo un subconjunto aleatorio de features en cada división.

Durante el entrenamiento, el algoritmo construye cada árbol de forma independiente dividiendo recursivamente el dataset en subconjuntos más pequeños basándose en las features que mejor separan los valores del target. Para esto utiliza nodos que representan decisiones sobre las features y hojas que representan las predicciones finales. En cada nodo, selecciona la feature y el umbral que minimizan la impureza (generalmente usando el error cuadrático medio o la varianza).

El proceso de bootstrap lo que hace es crear múltiples muestras aleatorias con reemplazo del dataset original, lo cual permite que cada árbol vea datos ligeramente diferentes y capture diferentes patrones. Esto ayuda a reducir la varianza del modelo y mejorar la generalización.

Durante la predicción, cada árbol hace su propia predicción y el modelo final promedia todas las predicciones de los árboles, lo cual reduce el error y hace al modelo más robusto ante outliers y ruido en los datos.

**Parámetros principales:**
- `n_estimators`: Número de árboles en el bosque. Más árboles generalmente mejoran el rendimiento pero aumentan el tiempo de entrenamiento
- `max_depth`: Profundidad máxima de cada árbol. Controla la complejidad del modelo y ayuda a prevenir sobreajuste
- `min_samples_split`: Mínimo de muestras requeridas para dividir un nodo interno. Valores más altos previenen divisiones sobre grupos muy pequeños de datos
- `min_samples_leaf`: Mínimo de muestras requeridas en una hoja. Penaliza hojas con muy pocas muestras, mejorando la generalización
- `max_features`: Número máximo de features a considerar en cada división. Valores como 'sqrt' o 'log2' ayudan a decorrelacionar los árboles

#### Búsqueda de hiperparámetros

Para hacer una búsqueda eficiente con Optuna creamos una función llamada objective_rf que recibe un objeto trial de Optuna que contiene sugerencias de hiperparámetros que introducimos, de forma a que construya un modelo RandomForestRegressor con esos hiperparámetros sugeridos (n_estimators, max_depth, min_samples_split, min_samples_leaf, max_features), y lo evalue mediante validación cruzada con KFold configurado con 5 folds. Considerando que Optuna minimiza por defecto el cross_val_score con scoring='neg_mean_absolute_error' que retorna valores negativos, retornamos el negativo del promedio para que Optuna minimice correctamente el MAE.

Para la optimización configuramos optuna.create_study con direction='minimize' para minimizar el MAE, y sampler=TPESampler(seed=RANDOM_STATE) que es el algoritmo de muestreo que utiliza modelos probabilísticos para sugerir mejores hiperparámetros basándose en los resultados anteriores. Luego ejecutamos study_rf.optimize con n_trials=200 que indica que probará 200 combinaciones diferentes de hiperparámetros, y show_progress_bar=True para visualizar el progreso de la búsqueda en tiempo real.

```python
def objective_rf(trial):
    params = {
    'n_estimators': trial.suggest_int('n_estimators', 50, 500),                         #numero de arboles en el ensemble
    'max_depth': trial.suggest_int('max_depth', 3, 30),                                 #profundidad maxima de cada arbol
    'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),                 #numero minimo de muestras requeridas para dividir un nodo interno
    'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 10),                   #numero minimo de muestras requeridas en una hoja
    'max_features': trial.suggest_categorical('max_features', ['sqrt', 'log2', None]),  #numero de features a considerar para cada division
    'random_state': RANDOM_STATE,
    'n_jobs': -1
}
    model = RandomForestRegressor(**params)
    
    cv = KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
    cv_scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='neg_mean_absolute_error', n_jobs=-1)
    
    return -cv_scores.mean()

study_rf = optuna.create_study(direction='minimize', sampler=TPESampler(seed=RANDOM_STATE))
study_rf.optimize(objective_rf, n_trials=500, show_progress_bar=True)
```

Luego de la búsqueda iterativa de parametros con Optuna que probó 500 hiperparámetros, el trial 111 fue el mejor consiguiendo un MAE medio de 16.466 en la validación cruzada con datos de train.

```python
best_rf = RandomForestRegressor(**study_rf.best_params, random_state=RANDOM_STATE, n_jobs=-1)
best_rf.fit(X_train, y_train)
print(f"\nMejor MAE (Random Forest): {study_rf.best_value:.4f}")
print(f"Mejores parámetros: {study_rf.best_params}")
```

### XGBoost

XGBoost es un algoritmo de regresión de boosting que lo que hace es construir secuencialmente múltiples modelos débiles (generalmente árboles de decisión) donde cada nuevo modelo corrige los errores del modelo anterior, mejorando gradualmente las predicciones. Para esto utiliza gradient boosting, que es una técnica que ajusta nuevos modelos para minimizar el gradiente de la función de pérdida.

Durante el entrenamiento, el algoritmo comienza con una predicción inicial (generalmente la media del target) y luego itera agregando árboles que corrigen los residuales del modelo anterior. Cada nuevo árbol se entrena sobre los residuales (errores) de los árboles anteriores, lo cual permite que el modelo capture patrones más complejos y mejore progresivamente.

El algoritmo utiliza regularización tanto en la función objetivo (L1 y L2) como en la estructura del árbol (gamma, min_child_weight), lo cual previene el sobreajuste y hace que el modelo sea más robusto. Además, utiliza técnicas como subsampling (tanto de filas como de columnas) que lo que hace es entrenar cada árbol con una muestra aleatoria del dataset y considerar solo un subconjunto de features, lo cual ayuda a reducir la varianza y mejorar la generalización.

Durante la predicción, el modelo suma las predicciones de todos los árboles, donde cada árbol contribuye con un peso determinado por la tasa de aprendizaje (learning_rate), lo cual permite un ajuste más fino y controlado del modelo.

**Parámetros principales:**
- `n_estimators`: Número de árboles (boosting rounds) a construir. Más árboles pueden mejorar el rendimiento pero también aumentar el sobreajuste
- `max_depth`: Profundidad máxima de cada árbol. Controla la complejidad individual de cada modelo débil
- `learning_rate`: Tasa de aprendizaje que controla qué tan rápido aprende el modelo. Valores más bajos requieren más árboles pero pueden mejorar la generalización
- `min_child_weight`: Peso mínimo requerido en una hoja. Valores más altos previenen hojas con muy pocas muestras
- `subsample`: Proporción de muestras a usar para entrenar cada árbol. Valores menores a 1.0 ayudan a prevenir sobreajuste
- `colsample_bytree`: Proporción de features a usar para entrenar cada árbol. Similar a max_features en Random Forest
- `gamma`: Reducción mínima de pérdida requerida para hacer una división. Valores más altos hacen el modelo más conservador
- `reg_alpha` y `reg_lambda`: Regularización L1 y L2 respectivamente que penalizan los valores de las hojas y previenen sobreajuste

#### Búsqueda de hiperparámetros

Para hacer una búsqueda eficiente con Optuna creamos una función llamada objective_xgb que recibe un objeto trial de Optuna que contiene sugerencias de hiperparámetros que introducimos, de forma a que construya un modelo XGBRegressor con esos hiperparámetros sugeridos (n_estimators, max_depth, learning_rate, min_child_weight, subsample, colsample_bytree, gamma, reg_alpha, reg_lambda), y lo evalue mediante validación cruzada con KFold configurado con 5 folds. Considerando que Optuna minimiza por defecto el cross_val_score con scoring='neg_mean_absolute_error' que retorna valores negativos, retornamos el negativo del promedio para que Optuna minimice correctamente el MAE.

Para la optimización configuramos optuna.create_study con direction='minimize' para minimizar el MAE, sampler=TPESampler(seed=RANDOM_STATE, n_startup_trials=30) que es el algoritmo de muestreo que utiliza modelos probabilísticos para sugerir mejores hiperparámetros basándose en los resultados anteriores, y pruner=MedianPruner(n_startup_trials=15) que permite detener anticipadamente los trials que muestran un rendimiento pobre durante el entrenamiento. Luego ejecutamos study_xgb.optimize con n_trials=300 que indica que probará 300 combinaciones diferentes de hiperparámetros, y show_progress_bar=True para visualizar el progreso de la búsqueda en tiempo real.

```python
def objective_xgb(trial):
    params = {
    'n_estimators': trial.suggest_int('n_estimators', 100, 1000),                    #numero de arboles
    'max_depth': trial.suggest_int('max_depth', 3, 12),                              #profundidad maxima de cada arbol, controla la complejidad de los modelos debiles
    'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),      #tasa de aprendizaje
    'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),                #peso minimo de las muestras en una hoja
    'subsample': trial.suggest_float('subsample', 0.6, 1.0),                         #porcentaje de filas utilizado en cada arbol
    'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),           #porcentaje de features utilizado en cada arbol
    'gamma': trial.suggest_float('gamma', 0, 5),                                     #reduccion minima de la perdida requerida para hacer una nueva division
    'reg_alpha': trial.suggest_float('reg_alpha', 1e-4, 10, log=True),               #regularizacion L1 sobre los pesos de los arboles
    'reg_lambda': trial.suggest_float('reg_lambda', 1e-4, 10, log=True),             #regularizacion L2 sobre los pesos de los arboles
    'random_state': RANDOM_STATE,
    'n_jobs': -1
}
    
    model = XGBRegressor(**params)
    
    cv = KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
    cv_scores = cross_val_score(model, X_train, y_train,cv=cv,scoring='neg_mean_absolute_error',n_jobs=-1)
    
    return -cv_scores.mean()

study_xgb = optuna.create_study(direction='minimize',sampler=TPESampler(seed=RANDOM_STATE, n_startup_trials=30),pruner=MedianPruner(n_startup_trials=15))
start_time = time.time()
study_xgb.optimize(objective_xgb, n_trials=500, show_progress_bar=True)
xgb_time = time.time() - start_time
```

Luego de la búsqueda iterativa de parametros con Optuna que probó 500 hiperparámetros, el trial 440 fue el mejor consiguiendo un MAE medio de 16.2007 en la validación cruzada con datos de train.

#### Entrenar el modelo con los mejores parámetros encontrados

Luego entrenamos el modelo con los mejores hiperparámetros encontrados por Optuna.

```python
best_xgb = XGBRegressor(**study_xgb.best_params, random_state=RANDOM_STATE, n_jobs=-1)
best_xgb.fit(X_train, y_train)
print(f"\nMejor MAE (XGBoost): {study_xgb.best_value:.4f}")
print(f"Tiempo: {xgb_time:.2f}s")
print(f"Mejores parámetros: {study_xgb.best_params}")
```
