## Dataset: Pokédex

El dataset utilizado es "Data of 1015 Pokémons" de Kaggle, que funciona como una Pokédex estructurada, el cual incluye información de 1025 Pokémons con variables de identidad, tipos, estadísticas de combate, rasgos y extras.

#### Identidad
- `id` — número de Pokédex  
- `name` — nombre del Pokémon  
- `rank` — rareza/categoría (ordinary, legendary, mythical, sub-legendary, etc.)  
- `generation` — región/generación (generation-i, generation-ii, …)  
- `evolves_from` — evolución previa  

#### Tipos
- `type1` — tipo principal  
- `type2` — tipo secundario (puede ser `None`)  

#### Stats de combate
- `hp`, `atk`, `def`, `spatk`, `spdef`, `speed` — estadísticas base  
- `total` — suma total de stats  

#### Rasgos y extras
- `height`, `weight` — medidas físicas  
- `abilities` — habilidades  
- `desc` — descripción Pokédex

##### **Test de Normalidad Shapiro-Wilk**

Para entender mejor la distribución antes del modelado hacemos el test de normalidad de Shapiro-Wilk que verifica si una muestra proviene de una distribución normal. Para ello hacemos dos hipótesis:

- **H₀ (Hipótesis nula):** Los datos siguen una distribución normal
- **H₁ (Hipótesis alternativa):** Los datos NO siguen una distribución normal

Si el test de Shapiro-Wilk es cercano a 1, la distribución es más compatible con la normalidad y si es cercano a cero es menos compativle con la normalidad.

Por lo cual, si:
- p > 0.05: No se rechaza la hipótesis nula, ya que no podemos asumir normalidad.
- p ≤ 0.05: Se rechaza la hipótesis nula.

```python
for col in numeric_cols:
    data = dataset[col].dropna()
    stat, p_value = stats.shapiro(data)
    print(f"{col}: estadístico={stat:.4f}, p-valor={p_value:.4f}")
```

Para todas las variables el p-valor del test de Shapiro-Wilk es 0.05<, por lo tanto no se cumple el supuesto de normalidad, entonces rechazamos la hipótesis nula. Lo cual coincide con lo que hemos visualizado en los histogramas y boxplots. Para la clasificación con los algoritmos que utilizaremos no es un requisito que las distribuciones sean normales.

Generalmente los Pokémones evolucionados mejoran sus estadísticas a excepcion de algunos que no evolucionan como por ejemplo Mewtwo que es un Pokémon legendario. Para analizarlos mejor cree una variable binaria de evolución.

```python
dataset['evolves'] = (dataset['evolves_from'] != 'nothing').astype(int)
```

Y cree una funcion que determina la etaá evolucitva del Pokémon, para verificar si el Pokémon del que evoluciona, tiene una evolución también.

```python
def count_evolution_stage(row):
    """Determina la etapa evolutiva del Pokémon"""
    if row['evolves_from'] == 'nothing':
        return 'Base'
    else:
        parent = row['evolves_from']
        parent_data = dataset[dataset['name'] == parent]
        if len(parent_data) > 0 and parent_data.iloc[0]['evolves_from'] != 'nothing':
            return 'Stage 3'
        else:
            return 'Stage 2'
dataset['evolution_stage'] = dataset.apply(count_evolution_stage, axis=1)
```

Luego contamos la evolucion y los stages.

```python
evolution_counts = dataset['evolves'].value_counts()
stage_counts = dataset['evolution_stage'].value_counts()
```

#### Test de Mann-Whitney

A continuación realizamos el test de Mann-Whitney que es una prueba no paramétrica que compara dos grupos independientes, para poder determinar si hay o no diferencia entre las distribuciones, donde:

- **H0:** No hay diferencia entre las distribuciones
- **H1:** Hay diferencia significativa

```python
ordinarios = dataset[dataset['is_legendary'] == 0]
legendarios_df = dataset[dataset['is_legendary'] == 1]
for col in numeric_cols:
    stat, pvalue = stats.mannwhitneyu(ordinarios[col], legendarios_df[col], alternative='two-sided')
    significativo = "SÍ" if pvalue < 0.05 else "NO"
    print(f"{col.upper()}: p-value = {pvalue:.2e} | Significativo: {significativo}")
```

Para todas las estadísticas de combate el p-value es menor a 0.05 por lo tanto se rechaza la hipótesis nula y asumimos que hay una diferencia significativa entre las distribuciones, confirmando que las distribuciones entre Pokémons ordinarios y legendarios son estadísticamente diferentes. Lo cual confirma lo que visualizó en los histogramas superpuestos, boxplots y violin plots que los Pokémones legendarios tienen estadísticas de combates superiores a los Pokémones ordinarios. Por lo tanto, todas las estadísticas de combate podrían ser variables útiles para la clasificación binaria entre Pokémons ordinarios y legendarios.

#### Test Chi-cuadrado

El test de chi-cuadrado se utiliza para testear la independencia de dos variables categóricas donde:

- **H0:** No hay asociacion entre las variables categóricas
- **H1:** Hay asociación entre las variables categóricas

Para ello definimos las variables categóricas que vamos a testear con is_legendary:

```python
categorical_vars = ['type1', 'type2', 'generation']
```

```python
chi2_results = []
for var in categorical_vars:
    contingency_table = pd.crosstab(dataset[var], dataset['is_legendary'])
    chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)
    chi2_results.append({
        'Variable': var,
        'Chi2': chi2,
        'p-value': p_value,
        'Significativo': 'Sí' if p_value < 0.05 else 'No'
    })
chi2_df = pd.DataFrame(chi2_results).sort_values('Chi2', ascending=False)
print(chi2_df.to_string(index=False))
```

Para type2 el p-value es mayor a 0.05 por lo tanto no se rechaza la hipótesis nula, en cambio para type1 y generation el p-value es menor a 0.05 por lo tanto para estas se rechaza la hipótesis nula y se asume mediante el test de chi-cuadrado que tiene una asociación significativa con la variable principal y que estas podrían ser variables predictivas para la clasificación entre Pokémones ordinarios y legendarios.

```python
df['offensive_ratio'] = (df['atk'] + df['spatk']) / (df['total'] + 1)
df['defensive_ratio'] = (df['def'] + df['spdef'] + df['hp']) / (df['total'] + 1)
```

Agregamos más features de estadísticas dada las features que tenemos. Estadísticas máximas y mínimas, un rango de estadísticas ya que los legendarios suelen ser más balanceados o tener un stat muy alto, la deviación estándard y el conteo de estadísticas superiores a 100.

```python
stat_cols = ['hp', 'atk', 'def', 'spatk', 'spdef', 'speed']
df['max_stat'] = df[stat_cols].max(axis=1)
df['min_stat'] = df[stat_cols].min(axis=1)
df['stat_range'] = df['max_stat'] - df['min_stat']
df['stat_std'] = df[stat_cols].std(axis=1)
df['stats_above_100'] = (df[stat_cols] >= 100).sum(axis=1)# Conteo de est
```

También mejoramos features categóricas, ssobre si evoluciona o no en binario y sobre el si tiene o no el tipo secundario.

```python
if 'evolves' not in df.columns:
    df['evolves'] = df['evolves_from'].notna().astype(int)
if 'has_type2' not in df.columns:
    df['has_type2'] = df['type2'].notna().astype(int)
```

También incluimos features acerca del total por no evoluciona ya que los legendarios suelen tener estadísticas altas y no evolucionar.

```python
df['total_x_no_evolves'] = df['total'] * (1 - df['evolves'])
```

Y una feature que convierte el valor total del cada Pokémon en su posición percentil dentro de todos los Pokémones del dataset.

```python
df['total_percentile'] = df['total'].rank(pct=True)
```

### Función evaluate_model

Creamos una funcion llamada evaluate_model la cual utilizaremos para hacer las predicciones, calcular las métricas, medir los tiempos de entrenamiento e inferencia, imprimir los resultados y la matriz de confusión para cada uno de los modelos que desarrollaremos.

```python
def evaluate_model(model, X_train, X_test, y_train, y_test, model_name, already_trained=False):    
    #Tiempo de entrenamiento
    if not already_trained:
        start_train = time.time()
        model.fit(X_train, y_train)
        train_time = time.time() - start_train
    else:
        train_time = 0.0
    
    #Tiempo de inferencia
    start_inf = time.time()
    y_pred = model.predict(X_test)
    inference_time = time.time() - start_inf
    
    y_pred_train = model.predict(X_train)
    
    metrics = {
        'Accuracy (Train)': accuracy_score(y_train, y_pred_train),
        'Accuracy (Test)': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred, zero_division=0),
        'Recall': recall_score(y_test, y_pred, zero_division=0),
        'F1-Score': f1_score(y_test, y_pred, zero_division=0),
        'Tiempo Entrenamiento (s)': train_time,
        'Tiempo Inferencia (s)': inference_time
    }
    print(f"{'='*55}")
    print(f"   {model_name}")
    print(f"{'='*55}")
    for metric, value in metrics.items():
        if 'Tiempo' in metric:
            print(f"   {metric}: {value:.6f}")
        else:
            print(f"   {metric}: {value:.4f}")
    print(f"\nMatriz de Confusión:")
    cm = confusion_matrix(y_test, y_pred)
    print(f"   TN={cm[0,0]:3d}  FP={cm[0,1]:3d}")
    print(f"   FN={cm[1,0]:3d}  TP={cm[1,1]:3d}")
    
    return metrics, y_pred
```

### GridSearchCV

GridSearchCV es un método para buscar automáticamente la mejor combinación de hiperparámetros de un modelo de machine learning. La pregunta es cómo funciona realmente: lo que hace es realizar una búsqueda exhaustiva considerando un conjunto definido de hiperparámetros, donde prueba todas las combinaciones posibles y selecciona la que tiene el mejor score en la validación cruzada. Es bastante completo, ya que evalúa todo el espacio de búsqueda, lo cual también hace que sea más costoso computacionalmente hablando.

Para esto se define una rejilla o grid de posibles valores para cada hiperparámetro, por ejemplo, valores de C = [0.1, 1, 10] y kernel = ['linear', 'rbf'], y GridSearch prueba todas esas combinaciones posibles (para este caso serían 6 modelos distintos). Para cada combinación hace una validación cruzada y calcula una métrica elegida y, al final, elige la combinación que da mejor rendimiento según esa métrica.

Luego definimos las grillas de hiperparámetros para cada modelo; el motivo de cada parámetro se explica en el comentario junto a cada uno.

#### Regresión Logística

Para la búsqueda de hiperparámetros configuramos C que controla la fuerza de la regularización, y la regularización l1 y l2, que hace referencia a lasso y ridge que son técnicas de regularización para reducir el sobreajuste y mejorar la generalización. También configuramos class_weight='balanced' que ajusta los pesos de las clases durante el entrenamiento para manejar el desbalanceo.

```python
param_grid_lr = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],          #fuerza de la regularizacion
    'penalty': ['l1', 'l2'],                      #lasso y ridge como regularizacion
    'solver': ['liblinear'],                      #definimos un solver que sea compatible con lasso y ridge
    'class_weight': ['balanced', None]            #para que las clases esten balanceadas
}
start_grid_lr = time.time()
grid_lr = GridSearchCV(LogisticRegression(random_state=RANDOM_STATE, max_iter=10000),param_grid_lr, cv=cv, scoring='f1', n_jobs=-1, return_train_score=True)
grid_lr.fit(X_train, y_train)
time_grid_lr = time.time() - start_grid_lr
print("Mejores hiperparámetros encontrados:")
for param, value in grid_lr.best_params_.items():
    print(f"   {param}: {value}")
print(f"\nMejor F1-score en CV: {grid_lr.best_score_:.4f}")
print(f"Tiempo total de búsqueda: {time_grid_lr:.4f}s")
print(f"Combinaciones evaluadas: {len(grid_lr.cv_results_['params'])}")
```

En la validación cruzada nos dió un score de 0.91 en f1 que nos indica que tanto la precisión como el recall están bien balanceados y es consistente en los 5 folds.

### Optuna

  <img src="https://optuna.org/assets/img/optuna-logo.png" width="250">
</div>

Optuna es un framework de optimización que utiliza algoritmos de optimización bayesiana, específicamente TPE (Tree-structured Parzen Estimator). A diferencia de GridSearchCV que evalúa todas las combinaciones y RandomizedSearchCV que muestrea aleatoriamente, Optuna aprende de las evaluaciones anteriores para dirigir la búsqueda hacia regiones más prometedoras del espacio de hiperparámetros.

Para utilizar Optuna necesitamos entender sus componentes principales:

- **Trial:** representa una evaluación individual del modelo con una combinación específica de hiperparámetros.
- **Study:** es el objeto que gestiona todo el proceso de optimización, almacenando los resultados de cada trial.
- **Objective:** es la función que define el problema de optimización, la cual Optuna intentará maximizar o minimizar.

| Ventaja | Descripción |
|---------|-------------|
| **Eficiencia** | Encuentra buenos hiperparámetros con menos evaluaciones que GridSearchCV |
| **Adaptabilidad** | Ajusta la búsqueda basándose en resultados previos |
| **Pruning** | Puede detener evaluaciones que no son prometedoras |
| **Flexibilidad** | Soporta espacios de búsqueda complejos y condicionales |

#### Regresión Logística

Para la búsqueda de hiperparámetros con Optuna definimos una función objetivo donde el estudio va probando distintas combinaciones de parámetros, como C, el tipo de regularización (L1/L2) y otros hiperparámetros del modelo. Donde en cada trial, Optuna entrena el modelo con una combinación distinta, y evalúa el rendimiento usando una métrica por ejemplo, F1-Score con validación cruzada y se queda con el mejor resultado. A diferencia de GridSearch y RandomizedSearch, Optuna usa técnicas de optimización más inteligentes como TPE sampler para explorar el espacio de búsqueda de forma más eficiente, concentrándose en las zonas donde encuentra mejores resultados.

```python
def objective_lr(trial):
    C = trial.suggest_float('C', 1e-3, 1e3, log=True)
    penalty = trial.suggest_categorical('penalty', ['l1', 'l2'])
    class_weight = trial.suggest_categorical('class_weight', ['balanced', 'none'])
    model = LogisticRegression(C=C, penalty=penalty, solver='liblinear', class_weight=None if class_weight == 'none' else class_weight,random_state=RANDOM_STATE, max_iter=10000)
    scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='f1', n_jobs=-1)
    return scores.mean()
```

```python
start_optuna_lr = time.time()
study_lr = optuna.create_study(direction='maximize', study_name='LR_optimization')
study_lr.optimize(objective_lr, n_trials=100, show_progress_bar=True)
time_optuna_lr = time.time() - start_optuna_lr

print(f"\nMejor trial: #{study_lr.best_trial.number}")
print("Mejores hiperparámetros encontrados:")
for param, value in study_lr.best_params.items():
    print(f"   {param}: {value}")
print(f"\nMejor F1-score en CV: {study_lr.best_value:.4f}")
print(f"Tiempo total de búsqueda: {time_optuna_lr:.4f}s")
print(f"Trials evaluados: {len(study_lr.trials)}")
```

Luego utilizamos los mejores hiperparámetros encontrados por RandomizedSearchCV y evaluamos el modelo.

```python
best_params_lr = study_lr.best_params
best_lr_optuna = LogisticRegression(C=best_params_lr['C'], penalty=best_params_lr['penalty'], solver='liblinear',class_weight=None if best_params_lr['class_weight'] == 'none' else best_params_lr['class_weight'],random_state=RANDOM_STATE, max_iter=10000)
metrics_lr_optuna, y_pred_lr_optuna = evaluate_model(best_lr_optuna, X_train, X_test, y_train, y_test, "REGRESIÓN LOGÍSTICA (Optuna)")
metrics_lr_optuna['Tiempo Entrenamiento (s)'] = time_optuna_lr
all_results.append({'Modelo': 'Regresión Logística', 'Método': 'Optuna', **metrics_lr_optuna})
```

#### SVM

Para la búsqueda de hiperparámetros del SVM con Optuna definimos primero una función objetivo donde especificamos los rangos de búsqueda para C, los kernels ('rbf', 'linear', 'poly') y gamma. En cada trial, Optuna propone una combinación distinta de estos hiperparámetros, entrena el modelo con esa configuración y evalúa el desempeño usando una métrica como el F1-Score con validación cruzada. Con los resultados anteriores, Optuna va ajustando las siguientes combinaciones que prueba y se va enfocando en las zonas del espacio de búsqueda donde encuentra mejores resultados, haciendo la optimización más eficiente que una búsqueda completamente exhaustiva o aleatoria.

```python
def objective_svm(trial):
    C = trial.suggest_float('C', 1e-1, 1e3, log=True)
    kernel = trial.suggest_categorical('kernel', ['rbf', 'linear', 'poly'])
    class_weight = trial.suggest_categorical('class_weight', ['balanced', 'none'])
    params = {
        'C': C, 'kernel': kernel, 'probability': True,
        'class_weight': None if class_weight == 'none' else class_weight,
        'random_state': RANDOM_STATE
    }
    if kernel in ['rbf', 'poly']:
        params['gamma'] = trial.suggest_categorical('gamma', ['scale', 'auto', 0.001, 0.01, 0.1])
    if kernel == 'poly':
        params['degree'] = trial.suggest_int('degree', 2, 5)
    model = SVC(**params)
    scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='f1', n_jobs=-1)
    return scores.mean()
```

```python
start_optuna_svm = time.time()
study_svm = optuna.create_study(direction='maximize', study_name='SVM_optimization')
study_svm.optimize(objective_svm, n_trials=100, show_progress_bar=True)
time_optuna_svm = time.time() - start_optuna_svm

print(f"\nMejor trial: #{study_svm.best_trial.number}")
print("Mejores hiperparámetros encontrados:")
for param, value in study_svm.best_params.items():
    print(f"   {param}: {value}")
print(f"\nMejor F1-score en CV: {study_svm.best_value:.4f}")
print(f"Tiempo total de búsqueda: {time_optuna_svm:.4f}s")
print(f"Trials evaluados: {len(study_svm.trials)}")
```

Luego utilizamos los mejores hiperparámetros encontrados con Optuna con el que se obtuvo el mejor score en la validación cruzada y evaluamos el modelo.

```python
best_params_svm = study_svm.best_params
svm_params = {
    'C': best_params_svm['C'], 'kernel': best_params_svm['kernel'],
    'probability': True, 'random_state': RANDOM_STATE,
    'class_weight': None if best_params_svm['class_weight'] == 'none' else best_params_svm['class_weight']
}
if best_params_svm['kernel'] in ['rbf', 'poly'] and 'gamma' in best_params_svm:
    svm_params['gamma'] = best_params_svm['gamma']
if best_params_svm['kernel'] == 'poly' and 'degree' in best_params_svm:
    svm_params['degree'] = best_params_svm['degree']

best_svm_optuna = SVC(**svm_params)
metrics_svm_optuna, y_pred_svm_optuna = evaluate_model(best_svm_optuna, X_train, X_test, y_train, y_test, "SVM (Optuna)")
metrics_svm_optuna['Tiempo Entrenamiento (s)'] = time_optuna_svm
all_results.append({'Modelo': 'SVM', 'Método': 'Optuna', **metrics_svm_optuna})
```

#### Árbol de Decisión

Para la búsqueda de hiperparámetros del Árbol de Decisión con Optuna también configuramos max_depth, que controla la profundidad máxima del árbol limitando cuántos niveles de decisiones puede tener, min_samples_split y min_samples_leaf que son técnicas de regularización que representan el número mínimo de muestras que puede tener tanto el nodo y como una hoja, criterion con 'gini' y 'entropy' como medidas de impureza, y ccp_alpha como parámetro de poda que penaliza la complejidad del árbol. En este caso, Optuna no elige las combinaciones al azar ni recorre todo el espacio, sino que, mediante una función objetivo y algoritmos de optimización bayesiana, va proponiendo combinaciones de estos hiperparámetros en sucesivos trials, aprendiendo de los resultados anteriores para concentrarse en las regiones del espacio donde el modelo obtiene mejor rendimiento.

```python
def objective_dt(trial):
    max_depth = trial.suggest_int('max_depth', 2, 20)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 30)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 25)
    ccp_alpha = trial.suggest_float('ccp_alpha', 0.0, 0.05)
    criterion = trial.suggest_categorical('criterion', ['gini', 'entropy'])
    class_weight = trial.suggest_categorical('class_weight', ['balanced', 'none'])
    model = DecisionTreeClassifier(max_depth=max_depth, min_samples_split=min_samples_split,min_samples_leaf=min_samples_leaf, ccp_alpha=ccp_alpha,criterion=criterion,class_weight=None if class_weight == 'none' else class_weight,random_state=RANDOM_STATE)
    scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='f1', n_jobs=-1)
    return scores.mean()
```

```python
start_optuna_dt = time.time()
study_dt = optuna.create_study(direction='maximize', study_name='DT_optimization')
study_dt.optimize(objective_dt, n_trials=200, show_progress_bar=True)
time_optuna_dt = time.time() - start_optuna_dt

print(f"\nMejor trial: #{study_dt.best_trial.number}")
print("Mejores hiperparámetros encontrados:")
for param, value in study_dt.best_params.items():
    print(f"   {param}: {value}")
print(f"\nMejor F1-score en CV: {study_dt.best_value:.4f}")
print(f"Tiempo total de búsqueda: {time_optuna_dt:.4f}s")
print(f"Trials evaluados: {len(study_dt.trials)}")
```

Y finalmente utilizamos los mejores hiperparámetros encontrados con el que obtuvimos el mejor score en la validación cruzada y evaluamos el modelo.

```python
best_params_dt = study_dt.best_params
best_dt_optuna = DecisionTreeClassifier(
    max_depth=best_params_dt['max_depth'],
    min_samples_split=best_params_dt['min_samples_split'],
    min_samples_leaf=best_params_dt['min_samples_leaf'],
    ccp_alpha=best_params_dt['ccp_alpha'],
    criterion=best_params_dt['criterion'],
    class_weight=None if best_params_dt['class_weight'] == 'none' else best_params_dt['class_weight'],
    random_state=RANDOM_STATE
)
metrics_dt_optuna, y_pred_dt_optuna = evaluate_model(
    best_dt_optuna, X_train, X_test, y_train, y_test, 
    "ÁRBOL DE DECISIÓN (Optuna)")
metrics_dt_optuna['Tiempo Entrenamiento (s)'] = time_optuna_dt
all_results.append({'Modelo': 'Árbol de Decisión', 'Método': 'Optuna', **metrics_dt_optuna})
```
